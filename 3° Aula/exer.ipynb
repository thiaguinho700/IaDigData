```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("monitoramento_tempo.csv")
```

```python
df['data'] = pd.to_datetime(df['data'])
```

```python
# 1)
plt.figure(figsize=(15,9), facecolor='gray')
plt.plot(df['data'],df['humidade'])
plt.title("Ano X Humidade", fontsize=20)
plt.xlabel("Ano")
plt.ylabel("Humidade")
plt.show()
```

```python
# 2)
plt.figure(figsize=(10,9), facecolor='gray')
plt.plot(df['data'], df['velocidade do vento'])
plt.title("Ano X Velocidade Vento", fontsize=15)
plt.ylabel("Velocidade Vento")
plt.xlabel("Ano")
plt.show()
```

```python
plt.plot(df['humidade'])
fig = plt.figure(figsize=(15,5))
eixo = fig.add_axes([0,0,1,1])

eixo.axhline(max(df['humidade']),color='k',linestyle= '--') 
eixo.axhline(min(df["humidade"]),color='k', linestyle="--")

x1=df['data'][df['humidade'].idxmax()] 
y1 = max(df['humidade'])
x2=x1=df['data'][df['humidade'].idxmax()]
y2= max(df['temperatura'])-5
eixo.annotate("Máximo",xy=(x1,y1), fontsize=20,xytext=(x2,y2), arrowprops=dict(facecolor='k'))

eixo.annotate("Máximo",xy=(x1,y1),fontsize=20,xytext=(x2,y2), arrowprops=dict(facecolor='k'))
```

```python
# 9)

df.columns = df.columns.str.strip().str.lower()

df['data'] = pd.to_datetime(df['data'])

mask_period = (df['data'] >= '2016-05-01') & (df['data'] <= '2016-08-31')
df_period = df[mask_period]

mask_year = (df['data'] >= '2016-01-01') & (df['data'] <= '2016-12-31')
df_year = df[mask_year]

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=False)

ax1.plot(df_period['data'], df_period['temperatura'], color='blue')
ax1.set_title('Temperatura de Maio 2016 até Agosto 2016')
ax1.set_ylabel('Temperatura (°C)')
ax1.grid(True)

ax2.plot(df_year['data'], df_year['temperatura'], color='red')
ax2.set_title('Temperatura de 2016 (Jan até Dez)')
ax2.set_ylabel('Temperatura (°C)')
ax2.set_xlabel('Date')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dados_x = np.linspace(0,10,10) # Geras valroes com biblioteca numpy
dados_y = dados_x * dados_x # Gera valores multiplicando um dado pelo o outro
```

```python
plt.figure(figsize=(10,5), facecolor='gray')
Ax = plt.axes()
Ax.set_facecolor('white')
plt.plot(dados_x,dados_y, color='red', linewidth=2, marker='o', linestyle="dashed", markeredgecolor='black')
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.title("Exemplo Matplotlib", fontsize=12)
```

```python
plt.rcParams.update({'font.size':15}) # Define o tamanho da fonte
plt.figure(figsize=(20,20)) # Define o tamanho da figura
```

```python
linhas = 2 # Definir a quantidade de linhas
colunas = 2 # Definir a quantidade de colunas
plt.subplot(linhas, colunas,1) # Define o numero de linhas e colunas 
plt.plot(dados_x, 'r--')
plt.title("Grafico 1")
# segundo grafico
plt.subplot(linhas, colunas, 2)
plt.plot(dados_x*dados_y, 'b-.')
plt.title("Grafico 2")
# terceiro grafico
plt.subplot(linhas, colunas,3)
plt.plot(dados_x/2, 'g--')
plt.title("Grafico 3")
# quarto grafico
plt.subplot(linhas, colunas, 4)
plt.plot(dados_y*2, 'r--')
plt.title("Gr√°fico 4")
plt.savefig("Exemplo subplot.jpg", dpi=300, bbox_inches='tight', transparent=True)
```

```python
# Grafico de barras
Figura = plt.figure() # 
Grafico = Figura.add_axes([0,0,0.7,0.7])
Rotulos = ['A','B','D','C','E','F','G','H','I','J']
Grafico.bar(Rotulos, dados_x)
plt.title("Quantidade x categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```

```python
# Grafico de barras

Figura = plt.figure() # 
Grafico = Figura.add_axes([0,0,0.7,0.7])
Rotulos = ['A','B','D','C','E','F','G','H','I','J']
Grafico.bar(Rotulos, dados_x, color='blue')
Grafico.bar(Rotulos, dados_x*0.1, color='red')
plt.title("Quantidade x categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```

```python
# Grafico de barras

Figura = plt.figure() # 
Grafico = Figura.add_axes([0,0,0.7,0.7])
Rotulos = ['A','B','D','C','E','F','G','H','I','J']
Grafico.bar(Rotulos, dados_x,dados_x*0.1, color=['r','b'])

plt.title("Quantidade x categoria")
plt.xlabel("Itens")
plt.ylabel("Quantidade")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
# Importando a base de dados
df = pd.read_csv('monitoramento_tempov2.csv')
```

```python

df['data'] = pd.to_datetime(df['data'])
plt.plot(df['data'], df['temperatura'])
plt.savefig("tempoXtemperatura", dpi=300, bbox_inches='tight')
plt.xlabel("Ano")
plt.ylabel("Temperatura")
plt.legend("Temperatura",loc='upper right') # Colocando legenda no grafico
```

```python
linha = 3
coluna = 3
plt.figure(figsize=(30,30), facecolor='gray')
plt.subplot(linha, coluna,1)
plt.plot(df['data'],df['velocidade do vento'])

plt.subplot(linha, coluna,2)
plt.plot(df['data'],df['direção do vento'])

plt.subplot(linha, coluna,3)
plt.plot(df['data'],df['humidade'])
plt.savefig("data_vento_humidade.jpg",dpi=300, bbox_inches='tight', transparent=True)
```
